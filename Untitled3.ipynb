```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/content/Mall.csv')
```

```python
print(df.head())
print(df.info())


print(df.isnull().sum())
```

```python
sns.set(style="whitegrid")
sns.histplot(df['Genre'])
plt.title('Gender distribution')

plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
df['Genre'] = df['Genre'].map({'Male': 1, 'Female': 0})


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
from sklearn.cluster import KMeans
inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
labels = kmeans.fit_predict(df_scaled)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
components = pca.fit_transform(df_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(components[:, 0], components[:, 1], c = labels, cmap='viridis')
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.title('Визуализация результатов кластеризации')
plt.show()
```

```python
df['Cluster'] = labels

cluster_characteristics = df.groupby('Cluster').mean()

print(cluster_characteristics)
```

```python
cluster_comparison = cluster_characteristics.transpose()
cluster_comparison['Difference'] = cluster_comparison.max(axis=1) - cluster_comparison.min(axis=1)
print(cluster_comparison.sort_values(by='Difference', ascending=False))
```
